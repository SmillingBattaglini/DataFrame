```python
#2. Calcular la media, mediana, y desviación típica de ambas columnas.

import numpy as np

data = {'ID': ['L', 'M', 'X', 'J', 'V'],
        'Calorias': [420, 380, 390, 490, 300],
        'Tiempo': [60, 40, 75, 55, 45]}

# Convertir la data en una array
calorias = np.array(data['Calorias'])
tiempo = np.array(data['Tiempo'])

# Calcular la media, mediana y desviación estandar por la columana de calorias
calorias_mean = np.mean(calorias)
calorias_median = np.median(calorias)
calorias_std = np.std(calorias)

# Calcular la media, mediana y desviación estandar por la columana del tiempo
tiempo_mean = np.mean(tiempo)
tiempo_median = np.median(tiempo)
tiempo_std = np.std(tiempo)

# Mostrar resultados
print("Calorias column:")
print("Mean:", calorias_mean)
print("Median:", calorias_median)
print("Desviación Standar:", calorias_std)

print("\nTiempo column:")
print("Mean:", tiempo_mean)
print("Median:", tiempo_median)
print("Desviación Standar:", tiempo_std)

```

```python
# Incluir los nuevos datos en la dataframe
import pandas as pd
import numpy as np

data = {'ID': ['L', 'M', 'X', 'J', 'V'],
        'Calorias': [420, 380, 390, 490, 300],
        'Tiempo': [60, 40, 75, 55, 45]}

# Crear una nueva dataframe
df = pd.DataFrame(data)

# Calcular la mediana, promedio y desviasión standar
calorias_mean = np.mean(df['Calorias'])
calorias_median = np.median(df['Calorias'])
calorias_std = np.std(df['Calorias'])

# Calcular los valores por cada columna
tiempo_mean = np.mean(df['Tiempo'])
tiempo_median = np.median(df['Tiempo'])
tiempo_std = np.std(df['Tiempo'])

# Añadir los nuevos valores a la nueva dataframe
df['Calorias_mean'] = calorias_mean
df['Calorias_median'] = calorias_median
df['Calorias_std'] = calorias_std
df['Tiempo_mean'] = tiempo_mean
df['Tiempo_median'] = tiempo_median
df['Tiempo_std'] = tiempo_std

# Mostrar los resultados
df
```
