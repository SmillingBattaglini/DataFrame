```python
df_2 = pd.read_csv('state-areas.csv')
df_2  # Muestra el DataFrame

# Cambiar el nombre de la columna area(sq. mi) para multiplicar más adelante
df_2.columns = ['state', 'area']
df_2.columns
df_2
```

```python

# 2. Revise los datos cargados y realice las modificaciones necesarias para empezar a trabajar

df_3 = pd.read_csv('state-population.csv')
df_3     # Muestra el DataFrame

df_3.columns = ['state', 'ages', 'year', 'population'] #Cambiar el nombre de la columna State/Region para usarlo como clave
df_3.columns
df_3
```

```python
# 5. Forme un nuevo dataset llamado df que tenga la información de los 3 sets importados. Use las llaves necesarias.
# Unimos los DataFrames utilizando la columna "state"
# Se mezclan los datos de las dataframe usando el método Outer para unir los datos de las filas

df_merge = pd.merge(df_1, df_2, on='state', how='outer') 
df_merge = pd.merge(df_merge, df_3, on='state', how='outer')
df_merge

df_merge.fillna(value = 1, inplace=True) # Reemplazar datos perdidos o NaN con 1
df_merge
```

```python
# 4.Agrupe por state/region y ages, para luego agregar usando promedio, mínimo, máximo y contar

df_concat = pd.concat([df_1, df_2, df_3])
df_grouped = df_concat.groupby(['state', 'ages']).mean()
df_grouped


```

```python
# 6. Calcule la población por área para cada estado. Para esto agrupe por
# state/region y calcule el promedio de la población. Guarde dicha
# información en un data frame con las columnas 'state' y 'pop/area'

df_merge.groupby(['state'])['population'].mean()
df_merge

```

```python
# 6. Calcule la población por área para cada estado. Para esto agrupe por
# state/region y calcule el promedio de la población. Guarde dicha
# información en un data frame con las columnas 'state' y 'pop/area'

df_merge= df_merge.assign(Multiplicacion = df_merge['area'] * df_merge['population'])

# Multiplicar dos columnas, sacarle el promedio y guardar el resultado en una nueva columna llamada "Promedio"
df_merge['Promedio'] = (df_merge['Multiplicacion']).mean()
df_merge

```

```python
df_merge = df_merge.dropna(axis=1)
df_merge

```

```python
df_merge['pop/area'] = df_merge['population'] / df_merge['area']
state_groups = df_merge.groupby('state')['pop/area'].mean()

result_df = pd.DataFrame(
    {'state': state_groups.index, 'pop/area': state_groups.values})
result_df
```

```python
# 7. Finalmente, cree un dataframe que posea: El estado, el área, la población, la población por área 
# y como índice la abreviación del nombre del estado.

import pandas as pd

# Crear el nuevo dataframe
df_new = pd.DataFrame()

# Agregar las columnas necesarias
df_new['state'] = df_merge['state']
df_new['area'] = df_merge['area']
df_new['population'] = df_merge['population']
df_new['pop/area'] = df_merge['population'] / df_merge['area']

# Establecer el índice
df_new.set_index(df_merge['abbreviation'],inplace = True)

# Mostrar el nuevo dataframe
df_new
```
